## Nomor 1

```python
mycursor = mydb.cursor()
query = '''select f.title,c.name as category,f.length
from film f
join film_category using (film_id) join category c using (category_id)
where c.name = "Comedy"
order by length asc,title asc
limit 10;'''
mycursor.execute(query)
myresult = mycursor.fetchall()
columns  = mycursor.column_names
df3 = pd.DataFrame(myresult,columns = columns)
df3
```

## Nomor 2

```python
mycursor = mydb.cursor()
query = '''select c.name as category, count(f.film_id) as banyak_film, avg(f.rental_rate)
from film f
join film_category using (film_id) join category c using (category_id)
group by c.name
order by banyak_film desc;'''
mycursor.execute(query)
myresult = mycursor.fetchall()
columns  = mycursor.column_names
df3 = pd.DataFrame(myresult,columns = columns)
df3
```
